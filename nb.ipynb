# Dailycovid in Jupyter 

This is a program to get county level COVID mortality statistics, then display them in Jupyter.

To install just run `pip3 install dailycovid`.

```python
import os
from IPython.display import Image, display
from IPython.display import IFrame
```

# Help

```python
print(os.popen('dailycovid --help').read())
```

# Getting the data
Run `dailycovid --get-data` to get the source data. 

Run again if you come back later and want to update with the most recent data.

```python
os.system('dailycovid --get-data')
```

# New York City, NY

```python
STATE = 'NY'
COUNTY = 'NEW YORK CITY'
STATE_COUNTY = f'{STATE}-{COUNTY}'

print(os.popen(f'dailycovid --plot -sc "{STATE_COUNTY}"').read())
```

```python
output_dir = 'output-counties'
image_path = f'plots_{COUNTY}_{STATE}.png'.replace(' ', '-')

for file in os.listdir(output_dir):
    if file == image_path:
        out_path = os.path.join(output_dir, file)
        print(f'![{out_path}]({out_path})')        
        display(Image(filename=out_path, width=750, height=750))
```
